pdf of $\mathcal N(1, 4)$

$
f(x|\mu=1,\sigma^{2}=4)=\frac{1}{2\sqrt{2\pi}}e^{-\frac{(x-1)^2}{8}}
$

```
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.stats as ss

mu = 1
var = 4
sigma = math.sqrt(var)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)

plt.plot(x, ss.norm.pdf(x,mu,sigma))
plt.show()
```
